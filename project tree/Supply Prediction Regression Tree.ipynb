```python
# Import library and dataset
import pandas as pd
import numpy as np
```

```python
raw = pd.read_csv('MOCK_DATA_7days.csv')
```

```python
raw.describe()
```

```python
raw
```

```python
# cleaning dataset
# Adding an is_holiday field
import random
is_holiday = [random.randrange(0, 2, 1) for i in range(1000)] 
raw['is_holiday'] = np.array(is_holiday)
```

```python
# fill nan value to zero
raw['food_1'].fillna(0, inplace=True)
raw['food_2'].fillna(0, inplace=True)
raw['food_3'].fillna(0, inplace=True)
raw['food_4'].fillna(0, inplace=True)
```

```python
raw.head()
```

```python
# Checking is there any null values
tf = raw.drop(["is_holiday","id"], axis=1)
tf.isna().sum()
```

```python
# Groupby each food based on date
tr=tf.groupby('date').sum()
```

```python
tr['is_holiday'] = np.array([0,0,0,0,1,1,0])
```

```python
tr
```

```python
from sklearn.tree import DecisionTreeRegressor
import matplotlib.pyplot as plt
X = tr.drop(['food_1','food_2','food_3','food_4'], axis=1)
Y = tr.drop('is_holiday', axis=1)
RegT = DecisionTreeRegressor(max_depth=5)
RegT.fit(X,Y)
```

```python
RegT.predict(X.iloc[0:7,:])
```
